```python
#Step 1: Read a CSV 

import sqlite3
import pandas as pd

#Load CSV
df = pd.read_csv('train.csv')
```

```python
#Step 2: Clean the CSV

df.info()
df['Age'] = df['Age'].fillna(df['Age'].mean())
df.drop(columns=['Cabin','Embarked'])
```

```python
#Step3: Load into SQLite

# Connect to a database (creates it if it doesn't exist)
conn_1 = sqlite3.connect('titanic.db')

# Write DataFrame into the database as a table
df.to_sql('passengers', conn_1, if_exists='replace', index=False)

print("Done - titanic.db created")
```

```python
#Step4: Run SQL queries

#Survival rate by passenger class and gender
result = """
SELECT PClass, Sex, ROUND(AVG(Survived),2) Survival_rate
FROM passengers
GROUP BY PClass, Sex
"""
#ROUND rounds the final answer to the mentioned number of decimal places
result_df = pd.DataFrame(pd.read_sql(result,conn_1))
```

```python
#Step5: Export results as a new CSV

result_df.to_csv('RESULT.csv', index=False)
```
