Step 1 — Load and first look

```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv('train.csv')
#print(df.head(5))
print(df.info())
```

Understand the data

```python
df.describe()

#zero_count = (df['Survived'] == 0).sum() #count the number of zero values in Survived column
##one_count = (df['Survived'] == 1).sum() #count the number of one values in Survived column
#print(zero_count)
#print(one_count)
```

Check missing values

```python
print(df.isnull().sum())
```

Clean the missing values

```python
df.columns
df['Age'] = df['Age'].fillna(df['Age'].mean())
#df = df.drop(columns=['Cabin','Embarked'])
print(df['Sex'])
```

Did Men survive more than women? Which passenger class member survived the most?

```python
#print(df[['Survived','Sex']])
df.groupby('Sex')['Survived'].mean() # Survival rate by gender
df.groupby('Pclass')['Survived'].mean()# Survival rate by passenger class
df.groupby('Survived')['Age'].mean()# Average age of survivors vs non-survivors
```

Charts

```python
# Chart 1: Survival count
df['Survived'].value_counts().plot(kind='bar')
plt.title('Survival Count')
plt.show()
```

```python
# Chart 2: Age distribution
df['Age'].hist()
plt.title('Age Distribution')
plt.show()
```

```python
# Chart 3: Survival by gender
df.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
# Survival rate by passenger class

df.groupby('Pclass')['Survived'].mean().plot(kind='bar')
plt.title('Survival Rate by Passenger Class')
plt.show()
```
