```python
#Step 1 — Create a SQLite database and load Titanic data into it
import sqlite3
import pandas as pd

#Load CSV
df = pd.read_csv('train.csv')

# Connect to a database (creates it if it doesn't exist)
conn = sqlite3.connect('titanic.db') #conn_1 is the variable or open "phone line" to the database that lets you talk to it

# Write DataFrame into the database as a table
df.to_sql('passengers', conn, if_exists='replace', index=False)

print("Done - titanic.db created")
```

Breakdown of Each Component

df.to_sql(...): This is a built-in pandas method used to export your structural DataFrame data into a relational database table.

'passengers': This string dictates the exact name of the table where your data will be stored inside the database.

conn: This is your active database connection object (usually created beforehand using libraries like sqlite3 or SQLAlchemy). It acts as the bridge telling pandas exactly where the database is hosted.

if_exists='replace': This defines the behavior if a table named passengers already exists in your database. The 'replace' setting tells pandas to completely delete (drop) the old table and create a fresh one with your new data.

index=False: By default, pandas tries to save the DataFrame's row index numbers (0, 1, 2, etc.) as an actual column in your SQL table. Setting this to False prevents that, saving only your explicit data columns.

```python
#Step 2: Query it with SQL

query = "SELECT * FROM passengers limit 5"
result = pd.read_sql(query,conn) # read the sql from the conn database
print(result)
```

```python
#Step 3

#Survival count by gender

count_gender = """
SELECT Sex, SUM(Survived) total_survived, COUNT(Survived), AVG(Survived) survival_rate 
FROM passengers
GROUP BY Sex
"""
pd.read_sql(count_gender,conn)

#Passengers over 30 who survived
survive_over_30 = """
SELECT Name, Age, Sex
FROM passengers
WHERE (Age > 30) AND (Survived = 1)
GROUP BY Age
"""
pd.read_sql(survive_over_30,conn)

# Average fare by passenger class
avg_fare = """
SELECT PClass, AVG(Fare) as avg_fare
FROM passengers
GROUP BY PClass
"""
pd.read_sql(avg_fare,conn)

#STEP 4 - Close Connection
conn.close()
print("Connection to database closed successfully")
```
